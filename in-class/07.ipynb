```python
import numpy as np
import pandas as pd
import holoviews as hv
hv.extension('bokeh')
```

```python
census = pd.read_csv("acs2015_county_data.csv")
census.head()
```

```python
from holoviews import dim

pov_inc = census[['Poverty', 'IncomePerCap']]

xlab = hv.Dimension(('Poverty', 'Poverty rate'), unit='%')
ylab = hv.Dimension(('IncomePerCap', 'Income per capita'), unit='USD')
scatter = hv.Scatter(pov_inc, xlab, ylab)
scatter.opts(color='k', marker='s', size=2,width=600, height=400,
             title="Relation between Income and Poverty in US Counties")
```

```python
frequencies, edges = np.histogram(census[['Unemployment']], 30)
hist = hv.Histogram((edges, frequencies))
hist.opts(title="Distribution of Unemployment Rate Across US Counties", width=600, height=400, color = 'red')
```

```python
scatter+hist
```

```python
vdims = [('PrivateWork', 'Private Sector Workers'), ('PublicWork', 'Public Sector Workers')]
ds = hv.Dataset(census, ['State'], vdims)
```

```python
ds = ds.aggregate(function=np.mean)
ds
```

```python
from holoviews import opts

states = ['New York', 'New Jersey', 'California', 'Texas']

# Plot for Private Work
bars_private = ds.select(State=states).to(hv.Bars, ['State'], ['PrivateWork']).sort()
bars_private.opts(opts.Bars(width=800, height=400, tools=['hover'], xrotation=90, show_legend=False))

# Plot for Public Work
bars_public = ds.select(State=states).to(hv.Bars, ['State'], ['PublicWork']).sort()
bars_public.opts(opts.Bars(width=800, height=400, tools=['hover'], xrotation=90, show_legend=False))

# Combine the plots
layout = bars_private + bars_public
layout
```
